### NOTE
If $\sum_{i} ||A||^2$ is the average of the squared error norm,This quantity can be written as $\frac{1}{n} ||A||^2$
Where $||A||_F=\sqrt (\sum_{i}^m\sum_{j}^n A_{ij}^2) = \sqrt(Tr(A^{T}A))$
### LOCO(optics & dispersion)
The parameters in the used model is varied by LOCO to minimize the $ \chi^2 $: deviation between the model and measured orbit response matrices $C_{model}$ and $C_{measured}$. Where $\sigma_i$ is the measured noise level on the ith BPM (variance of input error). And $W_i$ is the weights diagonal matrix $=\frac{1}{ \sigma^2_{i}}$.


$$  \sum_{i} ||\epsilon||^2= \chi^2 = \chi^2_{orm} +  \chi^2_{disp} \Rightarrow \alpha_1 \chi^2_{orm} +  \alpha_2 \chi^2_{disp} $$

$ \alpha_1 = 1, \alpha_2 = 0 , \alpha_3 = 0 \Rightarrow $ only optics corrections
$ \alpha_1 = 0, \alpha_2 = 1, \alpha_3 = 0  \Rightarrow$ only dispersion corrections
$ \alpha_1 = 0, \alpha_2 = 0, \alpha_3 = 1  \Rightarrow $optics and dispersion corrections

Weighted Least Square Method

$$  \chi^{2}_w =  \sum_{i} ((\alpha_1 \Delta C[0: -2, :] + \alpha_2 \Delta C[ -2 ::, :]+ \alpha_3 \Delta C)-(\sum_{k}\alpha_1 \frac{\partial{C[0: -2, :]}}{\partial{g_k}}  \Delta g_k + \sum_{k} \alpha_2 \frac{\partial{C[ -2 ::, :]}}{\partial{g_k}}  \Delta g_k +\sum_{k} \alpha_3 \frac{\partial{C}}{\partial{g_k}}  \Delta g_k))^{2}W_{i}$$

$$  \chi^{2}_w =   ((\alpha_1 \Delta C[0: -2, :] + \alpha_2 \Delta C[ -2 ::, :]+ \alpha_3 \Delta C)-(\sum_{k}\alpha_1 \frac{\partial{C[0: -2, :]}}{\partial{g_k}}  \Delta g_k + \sum_{k} \alpha_2 \frac{\partial{C[ -2 ::, :]}}{\partial{g_k}}  \Delta g_k +\sum_{k} \alpha_3 \frac{\partial{C}}{\partial{g_k}}  \Delta g_k))^{T}W((\alpha_1 \Delta C[0: -2, :] + \alpha_2 \Delta C[ -2 ::, :]+ \alpha_3 \Delta C)-(\sum_{k}\alpha_1 \frac{\partial{C[0: -2, :]}}{\partial{g_k}}  \Delta g_k + \sum_{k} \alpha_2 \frac{\partial{C[ -2 ::, :]}}{\partial{g_k}}  \Delta g_k  + \sum_{k} \alpha_3 \frac{\partial{C}}{\partial{g_k}}  \Delta g_k))$$


$$  \chi^{2}_w =  (\alpha_1 \Delta C[0: -2, :] + \alpha_2 \Delta C[ -2 ::, :] + \alpha_3 \Delta C)^T (\alpha_1 \Delta C[0: -2, :] + \alpha_2 \Delta C[ -2 ::, :] + \alpha_3 \Delta C) W - (\alpha_1 \Delta C[0: -2, :] + \alpha_2 \Delta C[ -2 ::, :] + \alpha_3 \Delta C)^T (\alpha_1 \frac{\partial{C[0: -2, :]}}{\partial{g}}   +  \alpha_2 \frac{\partial{C[ -2 ::, :]}}{\partial{g}}+\alpha_3 \frac{\partial{C}}{\partial{g}}) W \Delta g

-  (\alpha_1 \frac{\partial{C[0: -2, :]}}{\partial{g}}  \Delta g +  \alpha_2 \frac{\partial{C[ -2 ::, :]}}{\partial{g}} +  \alpha_3 \frac{\partial{C}}{\partial{g}})^T (\alpha_1 \Delta C[0: -2, :] + \alpha_2 \Delta C[ -2 ::, :] + \alpha_3 \Delta C) W + (\Delta g^{T}\frac{\partial{C[0: -2, :]}^T}{\partial{g}}\frac{\partial{C[0: -2, :]}}{\partial{g}}\Delta g W\alpha_1)+ (\Delta g^{T}\frac{\partial{C[ -2 ::, :]}^T}{\partial{g}}\frac{\partial{C[ -2 ::, :]}}{\partial{g}}\Delta g W\alpha_2) + (\Delta g^{T}\frac{\partial{C}^T}{\partial{g}}\frac{\partial{C}}{\partial{g}}\Delta g W\alpha_3)$$


$$ \frac{\partial{\chi^{2}_w}}{\partial{\Delta g}} = - 2 (\alpha_1 \frac{\partial{C[0: -2, :]}}{\partial{g}}  \Delta g +  \alpha_2 \frac{\partial{C[ -2 ::, :]}}{\partial{g}} +  \alpha_3 \frac{\partial{C}}{\partial{g}}) W (\alpha_1 \Delta C[0: -2, :] + \alpha_2 \Delta C[ -2 ::, :] +  \alpha_3 \Delta C) + 2 (\alpha_1 \frac{\partial{C[0: -2, :]}}{\partial{g}}  \Delta g +  \alpha_2 \frac{\partial{C[ -2 ::, :]}}{\partial{g}} + \alpha_3 \frac{\partial{C}}{\partial{g}})^{T} W (\alpha_1 \frac{\partial{C[0: -2, :]}}{\partial{g}}  \Delta g+  \alpha_2 \frac{\partial{C[ -2 ::, :]}}{\partial{g}} + \alpha_3 \frac{\partial{C}}{\partial{g}})\Delta g =0$$

$$ (\alpha_1 \frac{\partial{C[0: -2, :]}}{\partial{g}} +  \alpha_2 \frac{\partial{C[ -2 ::, :]}}{\partial{g}} +  \alpha_3 \frac{\partial{C}}{\partial{g}})^{T}(\alpha_1 \frac{\partial{C[0: -2, :]}}{\partial{g}} W \Delta g +  \alpha_2 \frac{\partial{C[ -2 ::, :]}}{\partial{g}} + \alpha_3 \frac{\partial{C}}{\partial{g}}) \Delta g= ( \alpha1 \frac{\partial{C[0: -2, :]}^{T}}{\partial{g}} + \alpha2 \frac{\partial{C[ -2 ::, :]}^{T}}{\partial{g}} + \alpha3 \frac{\partial{C}^{T}}{\partial{g}}) W(\alpha_1 \Delta C[0: -2, :] + \alpha_2 \Delta C[ -2 ::, :]+  \alpha_3 \Delta C)$$


$$  \Delta g =( (\alpha1 \frac{\partial{C[0: -2, :]}}{\partial{g}} ^{T}W\frac{\partial{C[0: -2, :]}}{\partial{g}})+(\alpha2 \frac{\partial{C[ -2 ::, :]}}{\partial{g}} ^{T}W\frac{\partial{C[ -2 ::, :]}}{\partial{g}}) + (\alpha3 \frac{\partial{C}}{\partial{g}} ^{T}W\frac{\partial{C}}{\partial{g}}))^{-1} ( \alpha1 \frac{\partial{C[0: -2, :]}^{T}}{\partial{g}} + \alpha2 \frac{\partial{C[ -2 ::, :]}^{T}}{\partial{g}} + \alpha3 \frac{\partial{C}^{T}}{\partial{g}}) W(\alpha_1 \Delta C[0: -2, :] + \alpha_2 \Delta C[ -2 ::, :]+ \alpha_3 \Delta C) $$




Differences between the matrices arise from quadrupole gradient differences between the model and real storage ring as well as differences in the BPM gains and steering magnet calibrations. Therefore the gains and steerer calibrations are also fit in the Ï‡2 minimization. The result is a model that best reproduces the measured response matrix. The model also accurately reproduces the beta functions, magnet gradients, BPM gains, and steering magnet calibrations of the real storage ring.
